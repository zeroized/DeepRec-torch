```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from example.loader.ctiteo_loader import load_data,missing_values_process
data=load_data('G:/dataset/criteo-small/train_1m.txt')
data=missing_values_process(data)
data.head()
```

```python
counts=data['I1'].value_counts()
counts=counts.sort_index()
counts_ratio=counts/len(data)
```

```python
total_acc=0
acc=0
idx=0
left=0
right=0
for i in range(len(counts_ratio)):
    total_acc+=counts_ratio[i]
    acc+=counts_ratio[i]
    print(i,counts_ratio[i])

    if acc>0.03:
        right=i
        print(left,right,idx,acc)
        idx+=1
        left=i+1
        acc=0
        if total_acc>0.97:
            print(total_acc)
            break
```
